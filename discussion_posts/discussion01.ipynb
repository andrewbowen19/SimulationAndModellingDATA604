```python
import numpy as np
import matplotlib.pyplot as plt
```

Let's write a function to simulate flipping one coin

```python
def flip_coin():
    """Simulates a single coin flip. Returns Heads/Tails with a 50/50 chance"""
    x = np.random.rand()
    return "Heads" if x > 0.5 else "Tails"
```

```python
# Simulate 1000 coin flips
results = []
for i in range(1000):
    coin = flip_coin()
    results.append(coin)
```

```python
results
```

Let's plot the count of our results (Heads or Tails). We'd expect by the law of large numbers that these should be roughly even assuming our coin flip functino is fair

```python
plt.hist(results)
plt.show()
```

We can plot our "cumulative results" with each roll below. This is done by treating a "heads" roll as a value of 1 and a "tails" value as a -1. The running total does diverge from 0 at points, but eventually returns to a value closer to 0 (though not exactly). If we re-ran our simulation above, we may get a different looking running total plot

```python
# Calculate running total of heads flipped
index = [i for i in range(0, len(results))]
cumulative_heads = 0
cum_running_total = []

# Iteratively calculate running total of heads (+1) and tails (-1)
for r in results:
    if r == "Heads":
        cumulative_heads += 1
    else:
        cumulative_heads -= 1
    cum_running_total.append(cumulative_heads)

# Plot running total
plt.plot(index, cum_running_total)
plt.xlabel("number of flips")
plt.ylabel("Cumulative total")
```
